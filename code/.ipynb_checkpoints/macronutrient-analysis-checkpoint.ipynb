## Add features

```python
# Add feature: number of meals logged, number of unique foods logged

def count_foods(x):
    count = 0
    for meal in x:
        count += len(meal['dishes'])
    return count

data['meal_count'] = data.food_entries.apply(lambda x: json.loads(x)[-1]['sequence'])
data['item_count'] = data.food_entries.apply(lambda x: count_foods(json.loads(x)))

goaldata = data.drop(columns=['food_entries', 'intakes_goals'])
```

```python
# Drop users with any NaN main macronutrients
#goaldata.loc[goaldata.isna().any(axis=1)].groupby('user_id').count()
drop_users = goaldata.loc[goaldata.isna().any(axis=1)].user_id.unique()
print(f'Num users dropped: {len(drop_users)}')
drop = goaldata[goaldata.user_id.isin(drop_users)].index
goaldata.drop(index=drop, inplace=True)
```

```python
# Relative % of calories
goaldata['sum_cals'] = goaldata['carb_total']*4 + goaldata['protein_total']*4 + goaldata['fat_total']*9

goaldata['protein_pc'] = goaldata['protein_total']*4/goaldata['sum_cals'] * 100
goaldata['fat_pc'] = goaldata['fat_total']*9/goaldata['sum_cals'] * 100
goaldata['carb_pc'] = goaldata['carb_total']*4/goaldata['sum_cals'] * 100

# Filter so data includes best tracked macronutrients -- ONLY FOR SINGLE DAY ANALYSIS
num = goaldata.shape[0]
num2 = (goaldata.loc[(goaldata['calorie_total'] - goaldata['sum_cals']) < 200].shape[0])
print(f'Rows dropped: {num - num2}')
goaldata = goaldata.loc[(goaldata['calorie_total'] - goaldata['sum_cals']) < 200]
goaldata['protein_pc'] = goaldata['protein_total']*4/goaldata['sum_cals'] * 100
goaldata['fat_pc'] = goaldata['fat_total']*9/goaldata['sum_cals'] * 100
goaldata['carb_pc'] = goaldata['carb_total']*4/goaldata['sum_cals'] * 100
```

```python
# Calc proportion of successes for differen macronutrient diets
def calc_num_success(df, prm):
    # Select above/below thresh
    quant = int(df[prm[1]].quantile(prm[0]))
    if prm[0] < 0.5:
        t = df.loc[df[prm[1]] < quant]
    else:
        t = df.loc[df[prm[1]] > quant]
    
    print()
    print(prm[2])
    print(f'Num days: {t.shape[0]}')
    print(f'Num users: {t.user_id.nunique()}')
    print(f'Success Rate:  {t.success.sum() / len(t.success)}')

# Parameters
thresholds = (0.25, 0.75)
feat = ('fat_pc', 'fat_pc', 'carb_pc', 'carb_pc', 'protein_pc', 'protein_pc')
diet = ('Low Fat', 'High Fat', 'Low Carb', 'High Carb', 'Low Protein', 'High Protein')
labs = [(thresholds[i%2], feat[i], typ) for i, typ in enumerate(diet)]

for val in labs:
    calc_num_success(goaldata, val)
```

When Low and High are defined as <20th and >80th, High Protein diet had highest success rate of 23.47%, followed by High Fat at 23.17% and Low Carb at 22.48%. Low Fat had the lowest success rate.

```python
# Rolling window to calculate success rates for windows
feat = ('fat_pc', 'carb_pc', 'protein_pc',)
diet = ('Fat', 'Carb', 'Protein')

plt.figure(figsize=(10, 6))

size = 5
window=[0, 100]
highest_success = 0

for i, j in zip(feat, diet):
    left = window[0]
    right = left + size
    d = goaldata[[i, 'user_id', 'success']]

    suc_rate = []
    days = []
    users = []
    x_vals = []
    while right <= window[1]:
        t = d.loc[(d[i] > left) & (d[i] < right)]
        days.append(t.shape[0])
        users.append(t.user_id.nunique())
        suc = t['success'].sum()/len(t)
        suc_rate.append(suc)
        x_vals.append(left)
 
        left += 1
        right = left + size
    
    suc_rate = pd.Series(suc_rate)
    print()
    print(f'{j} - Best Success Rate: {suc_rate.max()}')
    print(f'Num days: {days[suc_rate.idxmax()]}')
    print(f'Left side: {x_vals[suc_rate.idxmax()]}')
    
    plt.plot(x_vals, suc_rate, label=j)

#plt.title('Success Rate of Various Macronutrient using Sliding Window')
plt.xlabel(f'Percent of Calories From Each Macronutrient')
plt.ylabel('Success Rate')
plt.legend()
#plt.savefig('figures/macronutrient_successes_notitle.png')
```

```python
# do consecutive logging days have more success??

# calc days passed since last logged meal
goaldata.sort_values('date', inplace=True)
goaldata['shifted_date'] = goaldata.groupby('user_id').date.shift(1)
goaldata['days_passed'] = goaldata.groupby('user_id').apply(lambda x: (x.date - x.shifted_date).dt.days).reset_index(level='user_id', drop=True)
```

```python
goaldata['streak_id'] = np.nan
```

```python
# label each streak within user

c = 0
def label(x):
    global c
    x1 = x.reset_index(drop=True)
    k = np.zeros(len(x1))
    for i in range(len(x1)):
        if x1[i] == 1: # if 1 day has passed then part of same streak
            k[i] = c
        elif x1[i] > 1:  # if >1 day passed then init new streak id
            c += 1
    c += 1
    return pd.Series(k, index=x.index)
        
        
goaldata['streak_id'] = goaldata.groupby('user_id').days_passed.apply(lambda x: label(x))
```
